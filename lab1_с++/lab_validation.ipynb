```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function(x):
    return np.sin(x) * np.power(x, 2) - 10
```

```python
a = -20
b = 3
```

```python
x = np.arange(a, b+1)
y = []
[y.append(function(i)) for i in range(a, b+1)]
print(y)
```

```python
min_x = -20
min_y = -375
plt.plot(x, y, color='gray')
plt.scatter(min_x, min_y, color='red')
print(min_x)
```
